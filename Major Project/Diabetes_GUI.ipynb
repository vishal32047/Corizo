```python
import ipywidgets as widgets
from IPython.display import display

# Load the model
model = joblib.load('/content/drive/MyDrive/Colab Notebooks/diabetes_model.sav')

# Create input widgets
pregnancies = widgets.FloatText(description='Pregnancies')
glucose = widgets.FloatText(description='Glucose')
blood_pressure = widgets.FloatText(description='BloodPressure')
skin_thickness = widgets.FloatText(description='SkinThickness')
insulin = widgets.FloatText(description='Insulin')
bmi = widgets.FloatText(description='BMI')
dpf = widgets.FloatText(description='DiabetesPedigreeFunction')
age = widgets.FloatText(description='Age')

# predict button
predict_button = widgets.Button(description='Predict')


output_label = widgets.Label()

#prediction function
def predict_function(button):
    input_data = [
        pregnancies.value, glucose.value, blood_pressure.value, skin_thickness.value,
        insulin.value, bmi.value, dpf.value, age.value
    ]
    result = model.predict([input_data])
    output_label.value = f"Prediction: {'Diabetic' if result[0] == 1 else 'Not Diabetic'}"

# Attach the predict function to the button click event
predict_button.on_click(predict_function)

# Arrange widgets in a layout
input_widgets = [pregnancies, glucose, blood_pressure, skin_thickness, insulin, bmi, dpf, age]
input_box = widgets.VBox(input_widgets)
form = widgets.VBox([input_box, predict_button, output_label])

# Display the form
display(form)
```

```python
import joblib
import gradio as gr

# Load the model
model = joblib.load('/content/drive/MyDrive/Colab Notebooks/diabetes_model.sav')

# Define the prediction function
def predict_diabetes(pregnancies, glucose, blood_pressure, skin_thickness, insulin, bmi, dpf, age):
    input_data = [[pregnancies, glucose, blood_pressure, skin_thickness, insulin, bmi, dpf, age]]
    result = model.predict(input_data)
    text_out = ""
    herf1 = "https://www.diabetes.org/nutrition"
    herf2 = "https://www.diabetes.org/fitness"
    links = {
        "Diabetes Diet": herf1,
        "Physical Activity": herf2
    }

    if result[0] == 1:
        text_out = "You are a Diabetic 😟"
    else:
        text_out = "You are not a Diabetic 😃"

    info = f"For information about Diabetes Diet and Physical Activity, consult a healthcare professional.\n Diabetes Diet : ({links['Diabetes Diet']})\n Physical Activity : ({links['Physical Activity']})"

    return f"{text_out}\n{info}"

# Create Gradio Interface
iface = gr.Interface(
    fn=predict_diabetes,
    inputs=["number", "number", "number", "number", "number", "number", "number", "number"],
    outputs="text",
    live=True,
    title = "Disease Prediction"
)

# Launch the interface
iface.launch()
```
